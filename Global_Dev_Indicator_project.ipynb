```python
# function to fill missing values

def fill_missing_values(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        elif df[col].dtype == 'float64':
            df[col].fillna(df[col].mean(), inplace=True)
    return df

filled_df = fill_missing_values(df)

# print the sum of null values for each column
print(filled_df.isnull().sum())
```

```python
# the minimum values of global_resilience_score and	global_development_resilience_index are negative values
# changing all negative values in global_resilience_score	and global_development_resilience_index to the mean of the columns

df['global_resilience_score'] = df['global_resilience_score'].apply(lambda x: df['global_resilience_score'].mean() if x < 0 else x)
df['global_development_resilience_index'] = df['global_development_resilience_index'].apply(lambda x: df['global_development_resilience_index'].mean() if x < 0 else x)
```

```python
# Confirm that the problem is regression since the target variable, human_development_index, is continuous (values typically range from 0 to 1).

print(df['human_development_index'].dtype)  # Should be float64
print(df['human_development_index'].describe())  # Check range and distribution
```

```python
# total values of human development index column that are above 1 and lower than 0

print(df[df['human_development_index'] > 1].shape[0])
print(df[df['human_development_index'] < 0].shape[0])
```

```python
from sklearn.impute import SimpleImputer
df.loc[(df['human_development_index'] < 0) | (df['human_development_index'] > 1), 'human_development_index'] = None
imputer = SimpleImputer(strategy='median')
df['human_development_index'] = imputer.fit_transform(df[['human_development_index']])
print("HDI after imputation:")
print(df['human_development_index'].describe())
```

```python
#  importing regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# feature selection

X = df.drop(['human_development_index'], axis=1)
y = df['human_development_index']

# Identify categorical columns (object dtype)
categorical_cols = X.select_dtypes(include='object').columns

# Apply one-hot encoding to the categorical columns
X = pd.get_dummies(X, columns=categorical_cols, dummy_na=False) # dummy_na=False to drop NaN values

# SPLITTING DATA

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

# training the model with regression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# model evaluation

x_train_pred = model.predict(X_train)
x_test_pred = model.predict(X_test)

# A LINE GRAPH COMPARING PREDICTED AND ACTUAL VALUES

sns.lineplot(x=y_train, y=x_train_pred, label='Train')
sns.lineplot(x=y_test, y=x_test_pred, label='Test')
plt.legend()
```

```python
# checking for r square, mrse and mse

mse = mean_squared_error(y_test, x_test_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, x_test_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

TRAINING THE DATA WITH MULTIPLE MODELS TO SELECT THE BEST PERFORMING ONES

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=42),
    'LightGBM': LGBMRegressor(n_estimators=100, random_state=42, verbose=-1)
}

results = []
for name, model in models.items():
    # Train model
    model.fit(X_train_scaled, y_train)
    # Predict on test set
    y_pred = model.predict(X_test_scaled)
    # Calculate metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    results.append({'Model': name, 'RMSE': rmse, 'RÂ²': r2})
    print(f"{name} - Test RMSE: {rmse:.4f}, Test RÂ²: {r2:.4f}")

# Display results
results_df = pd.DataFrame(results)
print("\nModel Comparison:")
print(results_df)
```

HYPER PARAMETER TUNING

```python
from sklearn.model_selection import cross_val_score

for name, model in models.items():
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2')
    print(f"{name} - Cross-Validated RÂ² Mean: {scores.mean():.4f}, Std: {scores.std():.4f}")
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # Assuming pandas is needed for X.columns

tree_models = ['Random Forest', 'XGBoost', 'LightGBM']
for name in tree_models:
    model = models[name]
    # Ensure the model is fitted before accessing feature_importances_
    # This is a safeguard in case the previous fit wasn't executed
    # In a typical sequential notebook run, the fit in the previous cell is sufficient.
    # However, refitting here makes this cell independent.
    try:
        # Check if the model is already fitted (optional but good practice)
        # This check is not strictly necessary if you intend to refit anyway,
        # but can prevent unnecessary retraining.
        from sklearn.utils.validation import check_is_fitted
        check_is_fitted(model)
    except Exception: # Catch NotFittedError or similar exceptions
        print(f"Refitting model: {name}")
        model.fit(X_train_scaled, y_train) # Fit the model if not already fitted

    importances = model.feature_importances_
    indices = np.argsort(importances)[-10:]  # Top 10 features
    plt.figure(figsize=(8, 6))
    # Assuming X is still in scope and is a pandas DataFrame with column names
    plt.barh(range(len(indices)), importances[indices], align='center')
    plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
    plt.title(f"Top 10 Feature Importances - {name}")
    plt.xlabel("Importance")
    plt.tight_layout()
    plt.show()
```

```python
best_model = models['Random Forest']
y_pred = best_model.predict(X_test_scaled)
residuals = y_test - y_pred

plt.figure(figsize=(8, 5))
plt.scatter(y_pred, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residual Plot - Random Forest")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.tight_layout()
plt.show()
```
